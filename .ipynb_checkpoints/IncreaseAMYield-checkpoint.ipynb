# Python Packages

```python
# Import Python Packages

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import math # for trig on angles

import os # for reading csv files
import glob # for reading csv files
import re # to search strings for nums
```

# Constants

```python
# Constants used throughout the project

colors = {'blue':'#0b6374', 'orange':'#c0791b', 'gray':'#666666'} # colors use in presentation and graphs
tt_fs = 20 # plot title font size

layer_height = 0.03 # the layer height(mm) for the print was held constant

# Convert pixels to microns     
scale_factor = 29.2 # microns per pixel normal to print surface
incident_angle = 9.1 # degrees
pix_mic_x = scale_factor / math.cos(math.radians(incident_angle))
pix_mic_y = scale_factor
pix_mic_a = pix_mic_x * pix_mic_y

mpm_thresh_temps = {'t1':1600,'t2':1800,'t3':2000,'t4':2200} # temperature thresholds (C) used to calculate melt pool metrics
```

# Print Parameters (pp)

```python
# Read Print Parameters (PP) files
# settings used to print the part
# Calculate energy densities

pp = pd.read_csv('data/raw/pp/print_parameters.csv', index_col=0)
print('PRINT PARAMETERS')
print(pp)

# Calculate energy densities in pp df
# LED = linear laser energy density
# GED = global energy density
# VED = volumetric laser energy density

# calculate LED
pp['led'] = (pp['laser_power(W)'] / pp['scan_speed(mm/s)']).round(2) # Calculate LED = LP/SS

# calculate GED
pp['ged'] = (pp['laser_power(W)'] / \
    (pp['scan_speed(mm/s)'] * pp['hatch_spacing(mm)'])).round(2) # Calculate GED = LP/(SS*HS)

# calculate VED
pp['ved'] = (pp['laser_power(W)'] / \
    (pp['scan_speed(mm/s)'] * pp['hatch_spacing(mm)'] * layer_height )).round(2) # Calculate VED = LP/(SS*HS*LH)

print('PRINT PARAMETERS')
print(pp)
```

# Run Log (runs)

```python
# Read Run Log (runs) file
# Logs collected during print
# Create window type column
# Reorder columns

runs = pd.read_csv('data/raw/runs/runs.csv', index_col=0)
runs['date'] = pd.to_datetime(runs['date'])
print('RUN LOG')
print(runs.head())

# Create window type column in runs df based on window size (w,h)
for index, row in runs.iterrows():
    w = row['window_w(pixels)']
    h = row['window_h(pixels)']
    
    if w==384 and h==370:
        w_type = 'full'
    elif w==384 and h==80:
        w_type = 'quart_1'
    elif w==80 and h==370:
        w_type = 'quart_2'
    runs.loc[index,'window_type'] = w_type

# reorder runs columns
runs = runs[['date', 'condition', 'build_height(mm)', 'num_of_layers', 'frames_per_layer',
             'window_x(pixels)', 'window_y(pixels)',
             'window_w(pixels)', 'window_h(pixels)', 'window_type',
             'frame_rate(Hz)', 'exp_time(ms)', 'delay(ms)']]
print('RUN LOG')
print(runs.head())
```

# Melt Pool Metrics (mpm)

```python
# Read in mpm column labels

viz_col = list(pd.read_csv('data/raw/mpm/cols/viz_cols.csv', header=None).values.T.flatten()) # viz col headers
thresh_col = list(pd.read_csv('data/raw/mpm/cols/thresh_cols.csv', header=None).values.T.flatten()) # thresh col headers
```

```python
# Read in mpm tables viz (v) and threshold (t)
# Tables v and t are side by side tables with identical indexes, but different column names for different data

mpm_runs = {} # empty dictionary to read in raw mpm csv files
path = r'data/raw/mpm/data' # file path for csv files
all_files = glob.glob(os.path.join(path, "*.csv")) # clist of file names to be read
for f in all_files:
    run_nums = list(map(int, re.findall('\d+', f))) # search for #s in file name
    if 'viz' in f:
        vt = 'v'
        col_names = viz_col
    elif 'thresh' in f:
        vt = 't'
        col_names = thresh_col
    key = 'mpm_runs_%d-%d_%s' % (run_nums[1], run_nums[2], vt) # name dictionary key 
    mpm_runs[key] = pd.read_csv(f, names=col_names, header=None) # add df name to dict, read and add df as value, label columns
```

```python
# Combine mpm tables v and t, horizontally

mpm_temp = {} # empty dictionary to combine mpm(v/t) dfs
for x in list(mpm_runs.keys()): # run through keys of mpm_runs
    if 'v' in x: # only combine v with t, not t with v, otherwise there are duplicate dfs
        v_nums = list(map(int, re.findall('\d+', x))) # search for #s in df name
        for y in list(mpm_runs.keys()):
            t_nums = list(map(int, re.findall('\d+', y))) # search for #s in df name
            if v_nums == t_nums and 't' in y:
                combine_name = 'mpm_runs_%d-%d' % (v_nums[0], v_nums[1]) # name df, drop v and t notations
                mpm_runs.get(y).drop(['run', 'frame'], axis=1, inplace=True) # duplicate columns between v and t
                mpm_temp[combine_name] = pd.concat([mpm_runs.get(x), mpm_runs.get(y)], axis=1) # join v and t, and add to dict
mpm_runs = mpm_temp # set mpm_runs equal to the temporary df
```

```python
# Split mpm tables per run, vertically

mpm_temp = {} # empty dictionary to split dfs per run
for x in list(mpm_runs.keys()): # run through keys of mpm_runs
    run_nums = list(map(int, re.findall('\d+', x))) # search for #s in file name
    run_nums = range(run_nums[0], run_nums[-1]+1) # create a list of #s in the range 'Runs_n-m'
    for y in run_nums: # run through the list of run numbers
        key = 'mpm_run_%d' % (y) # name df per run
        mpm_temp[key] = mpm_runs.get(x)[mpm_runs.get(x)['run'] == y] #split df per run
mpm_runs = mpm_temp # set mpm_runs equal to the temporary df
```

```python
# Add columns to mpm for run num and condition num

cond_run = dict(zip(runs.index, runs.condition)) # create a diction from runs df, condition vs run

for x in list(mpm_runs.keys()): # run through keys of mpm_runs
    run_num = list(map(int, re.findall('\d+', x)))[0] # search for num in df_x name
    cond_num = cond_run[run_num] # set condition number based on run number of mpm_runs df
    mpm_runs.get(x)['run'] = run_num # create a run col in mpm_runs df, fill with a constant, run number
    mpm_runs.get(x)['condition'] = cond_num # create a condition col in mpm_runs df, fill with a constant, condition number
```

```python
# Combine all mpm tables into one large table

mpm = pd.concat(mpm_runs.values(), ignore_index=True) # combine add dfs from mpm_runs into a single df, vertically
mpm = mpm.set_index(['condition','run','layer','frame']) # set a multi-index in mpm
```

# Convert MPM Columns

```python
# Convert 'scan_direction(012)' into 'scan_direction(xy)'
# Update column labels
# Categorize the column

mpm = mpm[mpm['scan_direction(012)'] != 0] # delete all rows for 0/contours, only analyze internal melt pools
scan_dir_conv = {1:'x',2:'y'} # dictionary used to convert 012 into xy (x-axis, y-axis)
mpm['scan_direction(012)'].replace(scan_dir_conv, inplace=True) # use scan dict to convert 012 into xy
mpm.rename(columns={'scan_direction(012)':'scan_direction(xy)'}, inplace=True) # rename column with xy instead of 012
mpm['scan_direction(xy)'] = mpm['scan_direction(xy)'].astype('category') # set column type to category
```

```python
# Normalize intensities
# Update column labels

count_cols = [col for col in mpm.columns if 'counts' in col] # create a list of all columns that have the units 'counts'
for col in count_cols:
    mpm[col] = (mpm[col] / mpm['exp_time(ms)']).astype(int) # normalized intensity = intensity / exp_time, set to integer
    mpm.rename(columns={col:col.replace('counts', 'counts/ms')}, inplace=True) # remane intensity columns with new units
del mpm['exp_time(ms)'] # duplicate column of constant values already in runs df
```

```python
# Convert linear pixels to microns (length, width)
# Update column labels

lin_pix_cols = [col for col in mpm.columns if ('length' in col) or ('width' in col)] # linear pixel cols
for col in lin_pix_cols:
    mpm[col] = np.where(mpm['scan_direction(xy)']=='x', mpm[col] * pix_mic_x, mpm[col]) # scale value by x scale factor
    mpm[col] = np.where(mpm['scan_direction(xy)']=='y', mpm[col] * pix_mic_y, mpm[col]) # scale value by y scale factor
    mpm.rename(columns={col:col.replace('pixels', 'microns')}, inplace=True) # remane lin_pix columns
    lin_pix_cols[lin_pix_cols.index(col)] = col.replace('pixels', 'microns') # update the lin_pix_cols list
```

```python
# Convert area pixels to microns (area, sat_area)
# Update column labels

a_pix_cols = [col for col in mpm.columns if ('area' in col)] # area pixel cols  
for col in a_pix_cols:
    mpm[col] *= pix_mic_a # scale area by area scale factor
    mpm.rename(columns={col:col.replace('pixels', 'microns2')}, inplace=True) # remane a_pix columns
    a_pix_cols[a_pix_cols.index(col)] = col.replace('pixels', 'microns2') # update the a_pix_cols list
```

```python
# Offset pixel locations per window type (loc_x, loc_y)
# Convert pixels to microns
# Update column labels
# Currently set to "FALSE" so that the offsets and scales aren't taken into account
# When all the rows are averages per condition, the loc_x and loc_y columns will be meaningless data
# If there was a different analysis to look at per frame data, this would be used

loc_off_scale = False # if True, offset and scale pixels to microns

if loc_off_scale:
    loc_pix_cols = [col for col in mpm.columns if ('loc' in col)] # location pixel cols
    for col in loc_pix_cols:
        for index, row in mpm.iterrows():
            run_num = index[1] # get the run from the row
            w_type = runs.at[run_num,'window_type'] # use runs df with the run_num to get the window type

            # use the window type to align windows per window type, these are estimated offsets
            # because they are esimated, some offsets will cause loc_x and loc_y values to become negative
            if True: # if True, don't offset pixels to align windows per window type
                x_off = 0
                y_off = 0       
            elif w_type=='full':
                x_off = -20
                y_off = -20
            elif w_type=='quart_1':
                x_off = -22
                y_off = -13
            elif w_type=='quart_2':
                x_off = -5
                y_off = -24

            if 'loc_x' in col:
                mpm.at[index,'loc_x(pixels)'] = (mpm.at[index,'loc_x(pixels)'] + x_off) * pix_mic_x
            elif 'loc_y' in col:
                mpm.at[index,'loc_y(pixels)'] = (mpm.at[index,'loc_y(pixels)'] + y_off) * pix_mic_y
   
    # swap loc_x and loc_y cols and replace pixels with microns
    # the sensor_x_axis = printer_y_axis
    # the sensor_y_axis = printer_x_axis
    mpm.rename(columns={'loc_x(pixels)':'loc_y(microns)'}, inplace=True)
    mpm.rename(columns={'loc_y(pixels)':'loc_x(microns)'}, inplace=True)
```

```python
# For columns with microns or microns2 units, round to 2 decimal places

micron_cols = [col for col in mpm.columns if 'microns' in col] # list all columns that have units microns or microns2
for col in micron_cols:
    mpm[col] = mpm[col].round(2) # round to 2 decimal places
```

# Categorize MPM Columns

```python
# Reorder mpm cols by categories
# To reate multi-level columns, use keys when concatinating, for now, not needed, use above lists instead mpm[list]

# general data cols
if loc_off_scale: # offset and scale pixels to microns were computed above
    mpm_gen_cols = ['time(s)','loc_x(microns)','loc_y(microns)','scan_direction(xy)']
else: # offset and scale pixels to microns wasn't computed above
    mpm_gen_cols = ['time(s)','loc_x(pixels)','loc_y(pixels)','scan_direction(xy)']

# intensity cols
mpm_inten_cols = ['int_s_p(counts/ms)','int_l_p(counts/ms)',
                  'int_s_avg_3(counts/ms)','int_l_avg_3(counts/ms)',
                  'int_s_avg_5(counts/ms)','int_l_avg_5(counts/ms)',
                  'region_mean_short(counts/ms)','region_mean_long(counts/ms)'] 

# peak temperature cols
mpm_peak_cols = ['temp_hybref(C)','temp_peak(C)'] 

# threshold cols
mpm_thresh_df = pd.read_csv('data/raw/mpm/cols/thresh_cols_cat.csv', index_col=0) # t(1-4) and mpm_thresh_metrics
mpm_thresh_cols = []
for col in mpm_thresh_df.columns.values:
    mpm_thresh_cols += mpm_thresh_df[col].tolist() # convert df of col labels to list

# reorder mpm cols per category
mpm = pd.concat([mpm[mpm_gen_cols],
                 mpm[mpm_inten_cols],
                 mpm[mpm_peak_cols],
                 mpm[mpm_thresh_cols]],
                # keys=['general','intensity','peak','threshold'], # create a multilevel column df
                axis='columns')
print(mpm.info())
print(mpm.head(2))
```

# Average MPM per Condition

```python
# Average columns/metrics per rows/condition
# Remove columns where averaging is meaningless

cols_to_del = ['time(s)','loc_x(pixels)','loc_y(pixels)','scan_direction(xy)','temp_hybref(C)',
               't1_orient(degrees)','t2_orient(degrees)','t3_orient(degrees)','t4_orient(degrees)']

mpm_avg = mpm.drop(cols_to_del, axis=1) # delete meaningless columns when averaged

mpm_avg = mpm_avg.groupby('condition').mean().round(2) # average each column per condition, round to 2 decimal places
```

# Print Parameter Plot

```python
# Plot print parameters, comparing scan speed vs. laser power

pp_colors = {'PV':colors.get('blue'), 'SV':colors.get('orange')}

sns.set_style("ticks")

sns.lmplot('scan_speed(mm/s)', # Horizontal Axis
           'laser_power(W)', # Vertical Axis
           data=pp, # Data Source
           hue='cond_type', # Color Markers by
           palette=pp_colors, # Color Dictionary
           scatter_kws={'s': 100}, # Set Marker Size
           fit_reg=False, # Don't Fix a Regression Line
           legend=False # Remove Default Legend
          )

plt.title('PRINT PARAMETERS', fontsize=tt_fs)
plt.xlim(0, 4000)
plt.ylim(0, 400)
plt.legend(['Power Variation','Speed Variation'], title='Condition Type', loc='right', frameon=True)

# add vertical constant line for power with label
plt.axvline(x=1650, color=pp_colors.get('PV'), linestyle='dashed')
plt.text(1700, 90, '1650mm/s', fontsize=12, rotation=90)

# add horizontal constant line for speed with label
plt.axhline(y=370, color=pp_colors.get('SV'), linestyle='dashed')
plt.text(2400, 350, '370W', fontsize=12)

#plt.savefig('pp.jpeg', dpi=1000, bbox_inches="tight")
```
